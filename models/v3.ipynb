```python
import pandas as pd

games_df = pd.read_csv("../data/processed/v3.csv")
games_df
```

```python
labeled_df = games_df.drop(['win_name', 'lose_name', 'date'], axis=1)
labeled_df = labeled_df.rename(columns={'win_color': 'player_color', 'lose_color': 'opponent_color', 'win_rank': 'player_rank', 'lose_rank': 'opponent_rank'})
labeled_df['target'] = 1

labeled_df
```

```python
reversed_df = labeled_df.copy(deep=True)
reversed_df = reversed_df.rename(columns={'player_color':'opponent_color', 'opponent_color':'player_color', 'player_rank':'opponent_rank', 'opponent_rank':'player_rank', 'PS': 'OS', 'OS':'PS', 'PWR': 'OWR', 'OWR':'PWR'})
reversed_df['target'] = 0

reversed_df
```

```python
processed_df = pd.concat([labeled_df,reversed_df], ignore_index=True)
processed_df = processed_df[(~processed_df['player_rank'].isin(['Ama.', 'Insei', 'NR'])) & (~processed_df['opponent_rank'].isin(['Ama.', 'Insei', 'NR']))]
processed_df['player_rank'] = processed_df['player_rank'].apply(lambda x: int(x[0]))
processed_df['opponent_rank'] = processed_df['opponent_rank'].apply(lambda x: int(x[0]))
processed_df['player_color'] = processed_df['player_color'].apply(lambda x: 1 if x == 'b' else 0)
processed_df['opponent_color'] = processed_df['opponent_color'].apply(lambda x: 1 if x == 'w' else 0)

processed_df
```

```python
from sklearn.model_selection import train_test_split

X, y = processed_df[['player_color', 'opponent_color', 'player_rank', 'opponent_rank', 'komi', 'PWR', 'OWR', 'PS', 'OS','WWHR','WWHR-Error','LWHR','LWHR-Error', 'PNG', 'ONG']], processed_df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier()
clf.fit(X_train, y_train)

clf.score(X_test, y_test)
```

With new features: win rates and win streaks we got accuracy 3.7% increase.

Let's try with more stable rows. We can take a look by training with data with players who played more than 10 games

```python
processed_df = processed_df.loc[(processed_df['PNG']>10)|(processed_df['ONG']>10)]

processed_df
```

```python
X, y = processed_df[['player_color', 'opponent_color', 'player_rank', 'opponent_rank', 'komi', 'PWR', 'OWR', 'PS', 'OS','WWHR','WWHR-Error','LWHR','LWHR-Error', 'PNG', 'ONG']], processed_df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y)

clf = RandomForestClassifier()
clf.fit(X_train, y_train)

clf.score(X_test, y_test)
```

```python
import matplotlib.pyplot as plt
predictions = clf.predict(X_test)

plot_df = X_test.copy()
plot_df['actual'] = y_test
plot_df['prediction'] = predictions

plot_df['correct'] = plot_df.apply(lambda x: x['actual'] == x['prediction'], axis=1)
plot_df = plot_df.loc[plot_df['correct'] == False]
```

```python
colors = {True: 'green', False:'red'}
plt.scatter(plot_df['WWHR'], plot_df['LWHR'], color=plot_df['correct'].map(colors), s=1)
plt.show()
```

```python
colors = {True: 'green', False:'red'}
plt.scatter(plot_df['PWR'], plot_df['OWR'], color=plot_df['correct'].map(colors), s=1)
plt.show()
```

```python
from matplotlib import pyplot
from numpy import std
from pandas import Series

std = std([tree.feature_importances_ for tree in clf.estimators_], axis=0)
forest_importances = Series(clf.feature_importances_, index=['player_color', 'opponent_color', 'player_rank', 'opponent_rank', 'komi', 'PWR', 'OWR', 'PS', 'OS','WWHR','WWHR-Error','LWHR','LWHR-Error', 'PNG', 'ONG'])

fig, ax = pyplot.subplots()
forest_importances.plot.bar(yerr=std, ax=ax)
ax.set_title("Feature importances for Go Win Rates")
fig.tight_layout()
```
